```python
def THE_ULTIMATE_FUNCTION(array,n):
    BIG_VECTOR = []
    
    if array.all() == 0:
        Czl = np.zeros(len(array))
        Crl = np.zeros(len(array))
        return Czl,Crl
    else:
        for i in range(0,len(array)):
                str = array[i]
                arr = str.split()
                
                float_arr = []
                for i in arr:
                    
                    try:
                        I = float(i)
                        float_arr.append(I)
                    except:
                        continue 
                
                if len(float_arr) > 3:
                    composite_list = [float_arr[x:x+n] for x in range(0, len(float_arr),n)]

                    BIG_VECTOR.append(composite_list)
                    #print(composite_list)
                else:
                    BIG_VECTOR.append(float_arr)
                    #print(float_arr)          
        Czl =[]
        Crl = []

        for i in BIG_VECTOR:
            if len(i) == 1:
                Czl.append(np.nan)
                Crl.append(np.nan)
            elif len(i) >= 3:
                for j in i:
                    c = True 
                    while c == True: 
                        if type(j) == float:
                            Czl.append(i[2])
                            Crl.append(np.sqrt(i[0]**2+i[1]**2))
                            c = False 
                        elif len(j) == 3 :
                            Czl.append(j[2])
                            Crl.append(np.sqrt(j[0]**2+j[1]**2))
                            c = False
                        else:
                            c = False

        return Czl,Crl
```

```python
for i in ['Capture pi+ loc vector [mm]','Capture pi0 loc vector [mm]', 'Capture N16 loc vector [mm]',
 'Capture B12 loc vector [mm]',
 'Capture B13 loc vector [mm]',
 'Capture C10 loc vector [mm]',
 'Capture neutron loc vector [mm]']:
    
        print(i)
        z15, r15 = THE_ULTIMATE_FUNCTION(df15[str(i)],3)

        z20, r20 = THE_ULTIMATE_FUNCTION(df20[str(i)],3)

        z25, r25 = THE_ULTIMATE_FUNCTION(df25[str(i)],3)

        z30, r30 = THE_ULTIMATE_FUNCTION(df30[str(i)],3)

        z35, r35 = THE_ULTIMATE_FUNCTION(df35[str(i)],3)

        plt.hist(z15,alpha = 0.8, histtype = 'step')
        plt.hist(z20,alpha = 0.8, histtype = 'step')
        plt.hist(z25,alpha = 0.8, histtype = 'step')
        plt.hist(z30,alpha = 0.8, histtype = 'step')
        plt.hist(z35,alpha = 0.8, histtype = 'step')

        plt.legend(('15GeV','20GeV','25GeV','30GeV','35GeV'))
        plt.xlabel('Z location of %s ' %str(i))
        plt.ylabel('count')
        plt.yscale('log')
        plt.show()

        plt.hist(r15,alpha = 0.8, histtype = 'step')
        plt.hist(r20,alpha = 0.8, histtype = 'step')
        plt.hist(r25,alpha = 0.8, histtype = 'step')
        plt.hist(r30,alpha = 0.8, histtype = 'step')
        plt.hist(r35,alpha = 0.8, histtype = 'step')

        plt.legend(('15GeV','20GeV','25GeV','30GeV','35GeV'))
        plt.xlabel('R location of %s' %str(i))
        plt.ylabel('count')
        plt.yscale('log')
        plt.show()
```
