```python
df15 = pd.read_csv(r"C:\Users\Manuel\Desktop\3yr Project\15GeV_1000_SK.csv")
df15 = df15.drop(columns=['Unnamed: 0'])
df20 = pd.read_csv(r"C:\Users\Manuel\Desktop\3yr Project\20GeV_1000_SK.csv")
df20 = df20.drop(columns=['Unnamed: 0'])
df25 = pd.read_csv(r"C:\Users\Manuel\Desktop\3yr Project\25GeV_1000_SK.csv")
df25 = df25.drop(columns=['Unnamed: 0'])
df30 = pd.read_csv(r"C:\Users\Manuel\Desktop\3yr Project\30GeV_1000_SK.csv")
df30 = df30.drop(columns=['Unnamed: 0'])
df35 = pd.read_csv(r"C:\Users\Manuel\Desktop\3yr Project\35GeV_1000_SK.csv")
df35 = df35.drop(columns=['Unnamed: 0'])


#train_df = pd.concat(frames)
df15.head()
```

```python

###  RUELLE-TAKENS METHOD ###
### TAke a monodimensional series of data, shift it and create a new dimetion with the shifted version ###


%matplotlib widget

dfs = [df15, df20, df25, df30, df35]
attractor = pd.concat(dfs)

x1 = np.array(attractor['Ave KE e- [MeV]'])


y1, z1 = np.zeros(len(x1)), np.zeros(len(x1))


for i in range(0,len(x1)):
    y1[i-1]= x1[i]
    z1[i-2]= x1[i]


ipv.quickscatter(x1,y1,z1,
                 size=0.5, marker="sphere")
```

```python
%matplotlib inline 

for i in dfs:
    plt.scatter(x1,y1,marker='o', s= 1.5)
    plt.scatter(x1,z1,marker=',', s= 1.5)
    plt.scatter(y1,z1,marker='.', s= 1.5)

    plt.xlabel('Original Serie')
    plt.ylabel('Shifted Series')
    plt.legend(('XYprojection','XZprojection','YZprojection'))
    plt.show()
```
